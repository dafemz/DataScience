```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Cargar los datos
v1 = load_iris()
df = pd.DataFrame(v1.data, columns=v1.feature_names)
df['species'] = pd.Categorical.from_codes(v1.target, v1.target_names)

# Gráfica de variables nominales (especies)
plt.figure(figsize=(10, 5))
sns.countplot(x='species', data=df)
plt.title('Cantidad de Flores por Especie')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.show()

# Gráfica de variables ordinales (longitud de pétalos en rangos)
df['petal length (cm)'] = pd.cut(df['petal length (cm)'], bins=[0, 1.5, 3, 5, 7], labels=['Corta', 'Media Corta', 'Media Larga', 'Larga'])
plt.figure(figsize=(10, 5))
sns.countplot(x='petal length (cm)', data=df)
plt.title('Distribución de Longitud de Pétalos (Clasificada)')
plt.xlabel('Longitud de Pétalos')
plt.ylabel('Cantidad')
plt.show()

# Gráfica de variables discretas (histograma de longitud de sépalos)
plt.figure(figsize=(10, 5))
plt.hist(df['sepal length (cm)'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Longitud de Sépalos')
plt.xlabel('Longitud de Sépalos (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Gráfica de variables continuas (gráfica de dispersión)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Gráfica de Dispersión: Longitud vs Ancho de Sépalos')
plt.xlabel('Longitud de Sépalos (cm)')
plt.ylabel('Ancho de Sépalos (cm)')
plt.legend(title='Especie')
plt.show()

```
