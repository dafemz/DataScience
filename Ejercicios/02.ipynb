```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Cargar los datos
v1 = load_iris()
df = pd.DataFrame(v1.data, columns=v1.feature_names)
df['species'] = pd.Categorical.from_codes(v1.target, v1.target_names)

# Grfica de variables nominales (especies)
plt.figure(figsize=(10, 5))
sns.countplot(x='species', data=df)
plt.title('Cantidad de Flores por Especie')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.show()

# Grfica de variables ordinales (longitud de ptalos en rangos)
df['petal length (cm)'] = pd.cut(df['petal length (cm)'], bins=[0, 1.5, 3, 5, 7], labels=['Corta', 'Media Corta', 'Media Larga', 'Larga'])
plt.figure(figsize=(10, 5))
sns.countplot(x='petal length (cm)', data=df)
plt.title('Distribucin de Longitud de Ptalos (Clasificada)')
plt.xlabel('Longitud de Ptalos')
plt.ylabel('Cantidad')
plt.show()

# Grfica de variables discretas (histograma de longitud de spalos)
plt.figure(figsize=(10, 5))
plt.hist(df['sepal length (cm)'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Longitud de Spalos')
plt.xlabel('Longitud de Spalos (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Grfica de variables continuas (grfica de dispersin)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Grfica de Dispersin: Longitud vs Ancho de Spalos')
plt.xlabel('Longitud de Spalos (cm)')
plt.ylabel('Ancho de Spalos (cm)')
plt.legend(title='Especie')
plt.show()

```
